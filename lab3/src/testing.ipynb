testing minimax algorythm with a/b

```python
import minimax
import time
import numpy as np

DEPTH_RANGE = 6
ITERATIONS = 5

no_ab_times = []
ab_times = []
for depth in range(DEPTH_RANGE):
    temp = []
    for _ in range(ITERATIONS):
        start_time = time.time()
        minimax.play_game(depth, depth, False)
        temp.append(time.time() - start_time)
    no_ab_times.append(np.mean(temp))

    temp = []
    for _ in range(ITERATIONS):
        start_time = time.time()
        minimax.play_game(depth, depth, True)
        temp.append(time.time() - start_time)
    ab_times.append(np.mean(temp))
```

```python
import matplotlib.pyplot as plt

plt.plot(range(DEPTH_RANGE), no_ab_times, label='without alpha-beta')
plt.plot(range(DEPTH_RANGE), ab_times, label='with alpha-beta')
plt.xlabel('Depth')  # Title for the x-axis
plt.ylabel('Time [s]')  # Title for the y-axis
plt.title('Algorythm execution time for different depths')
plt.legend()
plt.show()
```

```python
plt.semilogy(range(DEPTH_RANGE), no_ab_times, label='without alpha-beta')
plt.semilogy(range(DEPTH_RANGE), ab_times, label='with alpha-beta')
plt.xlabel('Depth')  # Title for the x-axis
plt.ylabel('Time [s]')  # Title for the y-axis
plt.title('Algorythm execution time for different depths')
plt.legend()
plt.show()
```

Win rate for different depths

```python
import minimax
import numpy as np
import matplotlib.pyplot as plt

DEPTH_RANGE = 4
ITERATIONS = 10

win_rates = np.zeros((DEPTH_RANGE, DEPTH_RANGE))

for row in range(DEPTH_RANGE):
    for col in range(DEPTH_RANGE):
        scores = []
        for i in range(ITERATIONS):
            score = minimax.play_game(row, col, True)
            scores.append(score)
        
        # Calculate win rate for player 1
        win_rate = np.mean(np.array(scores))
        win_rates[row, col] = win_rate
```

```python
# Plot the results as a heatmap
plt.imshow(win_rates, cmap="viridis", interpolation="nearest")
plt.title("Player 1's Win Rate")
plt.suptitle("Player 1's win rates for different depth combinations")
plt.colorbar(label="Win Rate")
plt.tight_layout()

# Add labels for axes
plt.xticks(range(DEPTH_RANGE))
plt.yticks(range(DEPTH_RANGE))

# Show the plot
plt.show()
```
